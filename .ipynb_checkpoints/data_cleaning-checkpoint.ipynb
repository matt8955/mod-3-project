```python
import os
import functions as f
import pandas as pd
import tqdm
```

open all pickled df that have been scraped to clean

```python
for pickle in tqdm.tqdm(os.listdir('./pickled_dfs')):
    df = pd.read_pickle(f'./pickled_dfs/{pickle}')
    df = f.reformat_columns(df)
    df = f.typecast_df(df)
    df = f.rearrange_cols(df)
    df = f.date_to_column(df)
    try:
        pd.DataFrame.to_pickle(df, f'pickled_dfs_clean/{pickle}')
    except:
        os.mkdir('pickled_dfs_clean/')
```
