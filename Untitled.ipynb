```python
import requests
from bs4 import BeautifulSoup
import json
import tqdm 
import pandas as pd
import time
import csv
```

combine and create a list of all companies listed on 3 major exchanges
pulled from nasdaq website 

```python
nasdaq = pd.read_csv('nasdaq.csv')
amex = pd.read_csv('amex.csv')
nyse = pd.read_csv('nyse.csv')
```

```python
companies = pd.concat([nasdaq, amex, nyse])
```

```python
symbols = list(companies.Symbol)
```

now scrape all the tickers

```python
url = 'https://cloud.iexapis.com'
version = 'stable'
id_ = 'stock'
symbol = 'aapl'
type_ = 'chart'
time = 'max'
key = 'pk_7f427566b5444796bb43885380537e62'
pathway = '/stock/{symbol}/chart/{range}/{date}'
```

```python
import time
```

```python
frame_list = []
fails = []
for symbol in tqdm.tqdm(symbols[0:5]):
    req = requests.get(f'{url}/{version}/{id_}/{symbol}/{type_}/{time}?token={key}')
    symbol = symbol.strip()
    if req.status_code == 200:
        data = json.loads(req.text)
        df = pd.DataFrame(data)
        df['symbol'] = symbol
        df.columns = map(str.lower, df.columns)
        frame_list.append(df.copy())
        time.sleep(5)
    else:
        fails.append(symbol)
```
